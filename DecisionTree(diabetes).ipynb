```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
irisDataset = pd.read_csv('data_diabetes.csv',header=0)
```

```python
irisDataset.head()
```

```python
print(irisDataset)
```

```python
irisDataset = irisDataset.drop(labels="BMI",axis=1)
irisDataset = irisDataset.drop(labels="DiabetesPedigreeFunction",axis=1)
irisDataset = irisDataset.drop(labels="Age",axis=1)
```

```python
print(irisDataset)
```

```python
irisDataset = irisDataset.to_numpy()
```

```python
print(irisDataset)
```

```python
dataTraining = np.concatenate((irisDataset[0:40, :], irisDataset[50:90, :]),
                              axis=0)
dataTesting = np.concatenate((irisDataset[40:50, :], irisDataset[90:100, :]),
                              axis=0)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
inputTraining = dataTraining[:, 0:5]
inputTesting = dataTesting[:, 0:5]
labelTraining = dataTraining[:, 5]
labelTesting = dataTesting[:, 5]
print(labelTraining)
len(labelTraining)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar : ", prediksiBenar, "data")
print("Prediksi Salah : ", prediksiSalah, "data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah) *100,"%")
```
